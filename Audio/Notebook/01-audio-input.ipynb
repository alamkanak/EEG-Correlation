```python
import soundfile as sf
```

```python
# audio = AudioSegment.from_wav("New Recording 3.m4a")
data, samplerate = sf.read("window-from-3.flac")
```

```python
data.shape
```

```python
samplerate
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveplot(data, sr=samplerate)
```

```python
X = librosa.stft(data)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=samplerate, x_axis='time', y_axis='hz')
```

```python
import IPython.display as ipd
```

```python
ipd.Audio(data, rate=samplerate)
```

```python
import numpy as np
```

```python
data_fft = np.fft.rfft(data)
```

```python
data_fft.shape
```

```python
data.shape
```

```python
data_fft[:8]
```

```python
frequencies = np.abs(data_fft)
plt.plot(frequencies)
plt.title("FFT")
plt.xlim(0,4377)
plt.show()
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveplot(np.abs(np.fft.irfft(data_fft)), sr=samplerate)
```

```python
ipd.Audio(np.fft.irfft(data_fft), rate=samplerate)
```
